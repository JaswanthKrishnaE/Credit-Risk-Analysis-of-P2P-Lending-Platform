Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("pastel")
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings("ignore")
sns.set(style = 'whitegrid', font_scale = 1.1)
```

```python
filePath = "./Dataset/Bondora_preprocessed.csv"
df = pd.read_csv(filePath)
df.sample(5)
```

```python
df.Default.value_counts()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(data=df)
plt.xlabel("Numerical columns")
plt.xticks(rotation=90)
plt.show()
```

**UNI VARIATE ANALYSIS**

```python
sns.countplot(x = "Default", data=df, palette=["#003F7D", "#FD7702"])
# Set the title and axis labels
plt.title("Default Cases")
plt.xlabel("Default Conditions")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
plt.title("Heatmap of missing values")
plt.show()
```

```python
df.plot(subplots = True, figsize = (10,20))
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the histogram
sns.histplot(df.Age, bins=6, kde=True, color="#003F7D", edgecolor="#FFFFFF")

# Set the x-axis limits
plt.xlim(10, 80)

# Set the title and axis labels
plt.title("Age Distribution", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Add gridlines to the plot
plt.grid(False, axis="x")
plt.show()
```

```python
sns.barplot(x=df.Country.value_counts().values, y=df.Country.value_counts().index, color="#003F7D")
# Set the title and axis labels
plt.title("Count by Country")
plt.xlabel("Count")
plt.ylabel("Country")

# Show the plot
plt.show()
```

```python
sns.barplot(y=df.UseOfLoan.value_counts().values, x=df.UseOfLoan.value_counts().index, color="#003F7D")
# Set the title and axis labels
plt.title("Use of loan")
plt.xlabel("conditions")
plt.ylabel("count")
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
sns.barplot(y=df.MaritalStatus.value_counts().values, x=df.MaritalStatus.value_counts().index, color="#003F7D")
# Set the title and axis labels
plt.title("Marital Status Distribution")
plt.xlabel("Marital Status")
plt.ylabel("count")
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
sns.barplot(x=df.HomeOwnershipType.value_counts().values, y=df.HomeOwnershipType.value_counts().index, color="#003F7D")
# Set the title and axis labels
plt.title("Home Ownership Distribution")
plt.xlabel("Home Ownership Type")
plt.ylabel("count")
# Show the plot
plt.show()
```

```python
sns.countplot(x="Restructured", data=df, palette=["#003F7D", "#FD7702"])
# Set the title and axis labels
plt.title("Restructured")
plt.xlabel("Restructured")
plt.ylabel("Count")

# Show the plot
plt.show()
```

**Bi VARIATE ANALYSIS**

```python
df_numeric = df.select_dtypes(exclude='object')
df_numeric.sample(5)
```

```python
# Compute the correlation matrix
plt.figure(figsize=(15, 15))
corr_matrix = df_numeric.corr()
sns.heatmap(corr_matrix, annot=True, annot_kws={"size": 8})
# Set the title
plt.title("Bivariate Analysis: Correlation Heatmap")
# Show the plot
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

```python
# Create the countplot
sns.countplot(x="Default", data=df, hue="AgeGroup",palette="bright")
# Set the title and axis labels
plt.title("Default Cases")
plt.xlabel("Default Conditions")
plt.ylabel("Count")
# Show the plot
plt.show()
```

```python
# Create a scatter plot
sns.scatterplot(x="Age", y="Interest", data=df)

# Set the title and axis labels
plt.title("Bivariate Analysis: Age vs Interest")
plt.xlabel("Age")
plt.ylabel("Interest")

# Show the plot
plt.show()
```

```python
pivot_table = pd.pivot_table(data=df, index=['AgeGroup', 'NewCreditCustomer'], values='Interest', aggfunc='mean')

# Reset the index to have the columns available for plotting
pivot_table = pivot_table.reset_index()

# Create a bar plot
sns.barplot(x='AgeGroup', y='Interest', hue='NewCreditCustomer', data=pivot_table)

# Set the title and axis labels
plt.title('Average Interest by Age Group and New Credit Customer')
plt.xlabel('Age Group')
plt.ylabel('Average Interest')
plt.legend(loc='lower right')
# Show the plot
plt.show()
```

```python
# Create the pivot table
pivot_table = pd.pivot_table(data=df, index=['Gender', 'Restructured'], values=['Interest'], aggfunc='mean')

# Reset the index to have the columns available for plotting
pivot_table = pivot_table.reset_index()

# Create a bar plot
sns.barplot(x='Gender', y='Interest', hue='Restructured', data=pivot_table)

# Set the title and axis labels
plt.title('Average Interest by Gender and Restructured')
plt.xlabel('Gender')
plt.ylabel('Average Interest')
plt.legend(loc='lower right')

# Show the plot
plt.show()
```

```python
pivot_table = pd.pivot_table(data=df, index=['Gender','EmploymentStatus'],values ='Default' , aggfunc='count')
pivot_table = pivot_table.reset_index()

# Create the plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Default', hue='EmploymentStatus', data=pivot_table)

# Set the title and axis labels
plt.title('No. of Loans by Gender and Employment Status')
plt.xlabel('Gender')
plt.ylabel('No. of Loans')
plt.show()
```

```python
pivot_table = pd.pivot_table(data=df, index=['MaritalStatus', 'HomeOwnershipType'], values='Default', aggfunc='count')
pivot_table = pivot_table.reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='HomeOwnershipType', y='Default', hue='MaritalStatus', data=pivot_table)
plt.title('No, of loans by Marital Status and Home Ownership Type')
plt.xlabel('HomeOwnershipType')
plt.ylabel('No. of Loans')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Rating', y='NoOfPreviousLoansBeforeLoan')

# Set the title and axis labels
plt.title('Rating vs. Number of Previous Loans Before Loan')
plt.xlabel('Rating')
plt.ylabel('Number of Previous Loans Before Loan')

# Show the plot
plt.show()
```
